```python
#vamos a filtrar datos ya estadisticamente. Nos interesa saber el precio medio de venta de las viviendas 
#y determinar que caracteristicas es la mas demandada y a que precio de mercado se esta moviendo

sns.distplot(df["Precio"])
media =df["Precio"].mean()
std =df["Precio"].std()
print (media,std)
#Estrechamos el criterio a 1.5 para determinar con más precisión la media de precios.
plt.axvline( media, color="blue", linestyle='dashed')
plt.axvline( media +3*std, color= "red", linewidth=1)
plt.axvline( media-3*std, color="red", linewidth=1)

plt.show()
```

```python
#Para facilitar la busqueda de outlier definimos esta función

def buscar_outlier_std (df,columna,n):
    media =df[columna].mean()
    std =df[columna].std()
    return df[(df[columna]>media+3*std)|(df[columna]<media -3*std)]

buscar_outlier_std(df, "Precio",3)
```

```python
buscar_outlier_std(df, "m2",3)
```

```python
buscar_outlier_std(df,"Ano_construccion",3)
```

Procedemos a descartar los outlier_std de los indices. Para los año de construccion buscaremos el otro metodo de outlier, por cuantiles

```python
indice=buscar_outlier_std(df,"m2",3).index
df=df.drop(indice)
#df
```

Iremos a filtrar aun mas los años de construcción para eliminar o separar lo que podriamos considerar vivenda nueva o de segunda mano, el mezclar esos conceptos nos puede distorsionar la relacion precio vivienda de toda la tabla.

```python
sns.distplot(df["Ano_construccion"])
Q1 =df["Ano_construccion"].quantile(0.25)
Q3 =df["Ano_construccion"].quantile(0.75)
iqr= Q3-Q1
print (Q1,Q3,iqr)

plt.axvline( Q1, color= "red", linewidth=1)
plt.axvline( Q3, color="red", linewidth=1)
plt.axvline( Q1-iqr*0.45, color= "blue", linewidth=2)
plt.axvline( Q3+iqr*0.45, color="red", linewidth=2)
plt.show()
```

A pensar del estandar de 1.5, hemos decidido filtar un poco más para discriminar más los datos fuera de rango.

```python
indice=df[(df["Ano_construccion"]>Q3+iqr*0.45)|(df["Ano_construccion"]<Q1-iqr*0.45)].index
df=df.drop(indice)
```
